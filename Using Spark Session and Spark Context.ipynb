```
def isprime(n):
    n=abs(int(n))
    if n<2:
        return False
    elif(not n & 1):
        return False
    elif(n==2):
        return False
    for x in range(3,int(n**0.5)+1,2):
        if n%x==0:
            return False
    return True
```

```
nums = sc.parallelize(range(1000000))
```

```
print(nums.filter(isprime).count())
```

```
#(count the occurrences of unique words in a text line.)

# 1. Files loading  and defining the  functions in databricks.

text=sc.textFile("/FileStore/tables/demo.txt")
text
from operator import add
def tokenize(text):
     return text.split()

words = text.flatMap(tokenize)
words.collect()
```

```
wc = words.map(lambda x: (x,1))
wc.collect()
```

```
print(wc.toDebugString())
```

```
counts = wc.reduceByKey(add)
counts.collect() 
```

```
counts.saveAsTextFile("wcc")
```
